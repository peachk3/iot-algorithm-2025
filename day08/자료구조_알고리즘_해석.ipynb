### 1 Knapsack Problem

```python
# 1번 문제
def knapsack():
    array = [[0 for _ in range(maxWeight+1)] for _ in range(rowCount+1)]  # 2차원 배열 5x8
    for row in range(1, rowCount+1): 
        for col in range(1, maxWeight+1):
            if weight[row] > col:  # 이전값보다 컬럼(무게)가 작으면
                array[row][col] = array[row-1][col] # 이전값을 그대로 가지고 내려오고
            else: 
                value1 = money[row] + array[row-1][col-weight[row]]  # 최대무게에서 현재들어갈 무게를 뺀 위치의 가치+ 현재보석의 가격
                value2 = array[row-1][col]   # 이전위치 값
                array[row][col] = max(value1, value2) # 둘중에 비싼걸 
    
    return array[rowCount][maxWeight]   # 최종 행렬 마지막값(array[4][7])을 리턴

maxWeight = 7
rowCount = 4
weight = [0, 6, 4, 3, 5]
money = [0, 13, 8, 6, 12]

maxValue = knapsack()
print(f'베낭문제 최대 보석 가격 = {maxValue}')
```

### 2 유명맛집 대기줄

```python
# 2번
# 함수 선언
def isQueueFull():
    global SIZE, queue, front, rear

    if rear != SIZE - 1:
        return False
    elif rear == SIZE - 1 and front == -1:
        return True
    else:
        for i in range(front+1, SIZE):
            queue[i-1] = queue[i] # 데이터앞 빈자리에 첫번째 데이터 옮김
            queue[i] = None

        front -= 1
        rear -= 1
        return False

    
def isQueueEmpty():
    global SIZE, queue, front, rear
    if front == rear:
        return True
    else:
        return False
    
def enQueue(data):
    global SIZE, queue, front, rear
    if isQueueFull():
        print('Queue is full')
    else:
        rear += 1
        queue[rear] = data

def deQueue():
    global SIZE, queue, front, rear
    if isQueueEmpty():
        print('Queue is empty')
        return None
    else:
        front += 1
        data = queue[front]
        queue[front] = None
        return data

def peek():
    global SIZE, queue, front, rear
    if isQueueEmpty():
        print('Queue is empty')
        return None
    else:
        return queue[front+1]
    

# 초기화
SIZE = 5
queue = [None for _ in range(SIZE)]
front = rear = -1   

## 메인 모듈
if __name__ == '__main__':
    while True:
        select = input('삽입(I)/추출(E)/확인(V)/종료(Q) --> ').upper()

        if select == 'Q':
            break
        elif select == 'I':
            data = input('대기자 입력 > ')
            enQueue(data)
            print(f'대기줄 상태 : {queue}')
        elif select == 'E':
            data = deQueue()
            print(f'손닙입장 : {data}')
            print(f'대기줄 상태 : {queue}')
        elif select == 'V':
            data = peek()
            print(f'대기자명 : {data}')
            print(f'대기자 상태 : {queue}')
        else:
            print('입력 오류')
```

### 3 편의점 판매된 물건 목록 출력

```python
import random 

# TreeNode클래스 선언
class TreeNode:
    def __init__(self):
        self.left = None
        self.data = None
        self.right = None

# 초기화
memory = []
root = None
dataAry = [ '콘말차', '삿포로 맥주', '아카페라 벤티 헤이즐넛', '레어치즈푸딩', '척오리지널 메가 사워', '요아정 요거트컵', '페퍼로니피자주먹밥', 
            '널담 슈톨렌', '딸기마시멜로케이크', '버니공쥬 딸기뚱카롱', '고추잡채호빵', '체다 슈크림붕어스낵' ]
sellAry = [] # [random.choice(dataAry) for _ in range(20)]

print('오늘 판매된 물건 --> ', sellAry)

if __name__ == '__main__':
    while True:
        menu =  input('입력(i)/종료(q) > ').upper()
        if menu == 'Q':
            break
        elif menu == 'I':
            print('입력할 상품')  
            for idx, item in enumerate(dataAry):  # dataAry의 상품을 인덱스와 상품명을 디스플레이
                print(f'[{idx} | {item}]', end=',')
            print()
            pdtIdx = int(input('상품번호 > '))  # 인덱스를 입력하면
            sellAry.append(dataAry[pdtIdx])    # dataAry 인덱로 물건명을 sellAry에 추가가능


        
    node = TreeNode()
    node.data = sellAry[0]
    root = node
    memory.append(node)

    for name in sellAry[1:]: 
        node = TreeNode()
        node.data = name

        current = root
        while True:
            if name == current.data:
                break
            if name < current.data: # 현재 name이 노드안 데이터보다 작으면, 왼쪽으로 
                if current.left == None:
                    current.left = node
                    memory.append(node)
                    break # 연결했으니까 반복문 탈출
                else:
                    current = current.left # 왼쪽으로 더 내려감
            else: # 오른쪽으로 보냄
                if current.right == None:
                    current.right = node
                    memory.append(node)
                    break
                else:
                    current = current.right

print('이진 트리 구성 완료!')

# 전위순회 함수 정의
def preorder(node): # 루트노드를 파라미터로
    if node == None: return

    print(node.data, end='->')
    preorder(node.left) # 재귀호출(Recursive call)
    preorder(node.right) # 재귀호출

print('오늘 판매된 종류 ', end='->')
preorder(root)
```

### 4 SelfStudy 13-2

```python
import random 

## 이진 검색 함수 정의
def binSearch(ary, data):    
    pos = -1
    start = 0
    end = len(ary) - 1 # 리스트 마지막 인덱스

    while (start <= end):  # 시작이 끝보다 커지면 검색실패
        mid = (start + end) // 2
        if data == ary[mid]:
            return mid
        elif data > ary[mid]:
            start = mid + 1  # 뒤쪽에 데이터가 있다
        else:
            end = mid - 1 # 앞쪽에 데이터가 있다

    return pos


# 변수
dataAry = [random.randint(0, 100000) for _ in range(100000)]
dataAry.sort()   # 정렬을 1차로 해줘야(필수!!)
findData = random.randint(0, 100000)

print(f'배열 -> {dataAry}')
pos = binSearch(dataAry, findData)
if pos == -1:
    print(f'{findData} 없음')
else:
    print(f'{findData} / {pos}에 존재')
```
