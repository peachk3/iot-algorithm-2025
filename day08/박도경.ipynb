### 2번
- p.269 유명 맛집 대기줄 구현하기 문제를 while True:문을 사용하여 동적으로 입력받는 프로그램으로 변경 작성하시오.

```python
# 초기화
SIZE = 5
queue = [None for _ in range(SIZE)]
front = rear = -1

# 클래스, 함수 선언
def isQueueFull():
    # global - def 안에서만 전역변수의 값 바꿀 때 사용
    global SIZE, queue, front, rear
    # 맨 뒤의 데이터가 큐의 크기 -1과 동일할 때 (큐가 가득 참)
    if rear == (SIZE - 1):
        return True
    else:
        return False
    
def isQueueEmpty():
    global SIZE, queue, front, rear
    # 맨 앞의 데이터가 맨 뒤의 데이터와 동일 할 때 (큐가 비어있음)
    if front == rear:
        return True
    else:
        return False

def enQueue(data):
    global SIZE, queue, front, rear
    if isQueueFull():
        print('대기 마감')
        return
    # 맨 뒤의 데이터 다음에 data 삽입
    rear += 1
    queue[rear] = data

def deQueue():
    global SIZE, queue, front, rear
    if isQueueEmpty():
        print('대기 없음')
        return None
    front += 1
    data = queue[front]
    queue[front] = None

    # 앞의 인덱스에 값 입력하고 i의 인덱스에 None 입력
    for i in range(front + 1, rear + 1):
        queue[i - 1] = queue[i]
        queue[i] = None
    front = -1
    rear -= 1

    return data

def peek():
    global SIZE, queue, front, rear
    if isQueueEmpty():
        print('바로 입장')
        return None
    return queue[front + 1]

# 메인 함수
if __name__ == '__main__':
    while True:
        select = input('등록(I)/입장(E)/확인(V)/종료(Q)').upper()
        
        if select == 'Q':
            print('식당 영업 종료')
            break
        
        elif select == 'I':
            data = input('이름 입력 >')
            enQueue(data)
            print(f'대기 줄 상태 : {queue}')

        elif select == 'E':
            print(deQueue(), '님 식당에 들어감')
            print('대기 줄 상태 : ', queue)

        elif select == 'V':
            data = peek()
            print(f'대기 줄 상태 확인 : {queue}')
        
```

### 3번
- p.309 편의점 판매된 물건 목록 출력 -> while True:문으로 동적으로 입력받는 프로그램으로 변경하기
- 데이터 종류  [ 콘말차, 삿포로 맥주, 아카페라 벤티 헤이즐넛, 레어치즈푸딩, 척오리지널 메가 사워, 요아정 요거트컵, 페퍼로니피자주먹밥, 널담 슈톨렌, 딸기마시멜로케이크, 버니공쥬 딸기뚱카롱, 고추잡채호빵, 체다 슈크림붕어스낵 ]

```python
import random

# 함수 선언 부분
class TreeNode():
    def __init__(self):
        self.left = None
        self.data = None
        self.right = None
    
# 전역 변수 선언
memory = []
root = None
dataAry = []
# ['콘말차', '삿포로 맥주', '아카페라 벤티 헤이즐넛', '레어치즈푸딩', '척오리지널 메가 사워', 
#    '요아정 요거트컵', '페퍼로니피자주먹밥', '널담 슈톨렌',' 딸기마시멜로케이크', '버니공쥬 딸기뚱카롱', 
#    '고추잡채호빵', '체다 슈크림붕어스낵']

while True:
    select = input('판매 상품 입력(I)/판매 상품 확인(C)/ 판매 종류 확인(V)/종료(Q)').upper()
    
    if select == 'Q':
        print('편의점 영업 종료')
        break

    elif select == 'I':
        name = input('판매된 상품 > ') 
        dataAry.append(name)
        print(f'{name} 입력 완료', end=' ')

    elif select == 'C':
        sellAry = [random.choice(dataAry) for _ in range(20)]
        print(f'오늘 판매된 물건(중복 O) -> {sellAry} ', end='')

    elif select == 'V':
        # 루트노드 생성
        node = TreeNode()
        node.data = sellAry[0]
        root = node
        memory.append(node)

        for name in sellAry[1:]:
            node = TreeNode()
            node.data = name

            current = root
            while True:
                if name == current.data:
                    break
                if name < current.data:
                    if current.left == None:
                        current.left = node
                        memory.append(node)
                        break
                    current = current.left
                else:
                    if current.right == None:
                        current.right = node
                        memory.append(node)
                        break
                    current = current.right

        # 전위순회
        def preorder(node):
            # 루트노드를 파라미터로 사용
            if node == None:
                return
            print(node.data, end=' ')
            preorder(node.left) # 재귀호출
            preorder(node.right) # 재귀호출

        print('오늘 판매된 종류(중복 X) ->', end=' ')
        # 함수호출
        preorder(root)   
    print()
```

### 4번
- p.473 Self Study 13-2의 문제 -> 이진 검색 알고리즘을 풀면서 검색 획수도 같이 출력하도록 구현하시오.

```python
import random
count = 0

# 이진 검색 함수 정의
def binSearch(ary, data):
    global count
    pos = -1
    start = 0
    end = len(ary) - 1 # 리스트 마지막 인덱스

    while start <= end : # 시작이 끝보다 커지면 검색실패
        mid = (start + end) // 2
        if data == ary[mid]:
            count += 1
            return mid
        elif data > ary[mid]:
            start = mid + 1 # 뒤쪽에 데이터가 있다
            count += 1
        else:
            end = mid - 1 # 앞쪽에 데이터가 있다
            count += 1
    
    return pos 

findData = int(input('찾을 값을 입력(숫자)'))

# 변수 선언
# 랜덤 숫자 배열 출력
dataAry = [random.randint(0, 1000000) for _ in range(100000)]
dataAry.insert(random.randint(0, 1000000), findData)
dataAry.sort() # 배열 정렬

print(f'배열 일부 -> {dataAry}')
pos = binSearch(dataAry, findData)
if pos == -1:
    print(f'{findData} 없음')
else:
    print(f'{findData} / {pos}에 존재')
    print(pos)
print(f'{count}회 검색함')
```
